```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Fri Apr 24 11:25:38 2020

@author: bojanstavrikj
"""

from bs4 import BeautifulSoup as BS
from selenium import webdriver
from functools import reduce
import pandas as pd
import time
from selenium import webdriver
from kora.selenium import wd
import csv
import datetime
from google.colab import files

# function to load wunderground data (without this it has no records to show)
def render_page(url):
    #driver = webdriver.Chrome(ChromeDriverManager().install())
    wd.get(url)
    time.sleep(3)
    r = wd.page_source
    return r


# function to scrape wunderground
def scraper(page, dates):
    output = pd.DataFrame()

    for d in dates:

        url = str(str(page) + str(d))

        r = render_page(url)

        soup = BS(r, "html.parser")
        container = soup.find('lib-city-history-observation')
        check = container.find('tbody')

        data = []

        if(check == None):
            continue

        for c in check.find_all('tr', class_='ng-star-inserted'):
            for i in c.find_all('td', class_='ng-star-inserted'):
                trial = i.text
                trial = trial.strip('  ')
                data.append(trial)

        length = len(data)
        Time = pd.DataFrame([data[x] for x in range(0,length,10)], columns=['Time'])
        Temperature = pd.DataFrame([data[x] for x in range(1,length,10)], columns=['Temperature'])
        Dew_Point = pd.DataFrame([data[x] for x in range(2,length,10)], columns=['Dew_Point'])
        Humidity = pd.DataFrame([data[x] for x in range(3,length,10)], columns=['Humidity'])
        Wind = pd.DataFrame([data[x] for x in range(4,length,10)], columns=['Wind'])
        Wind_Speed = pd.DataFrame([data[x] for x in range(5,length,10)], columns=['Wind_Speed'])
        Wind_Gust = pd.DataFrame([data[x] for x in range(6,length,10)], columns=['Wind_Gust'])
        Pressure = pd.DataFrame([data[x] for x in range(7,length,10)], columns=['Pressure'])
        Precipitation = pd.DataFrame([data[x] for x in range(8,length,10)], columns=['Precipitation'])
        Condition = pd.DataFrame([data[x] for x in range(9,length,10)], columns=['Condition'])
        print(str(str(d) + ' finished!'))
        dfs = [Time, Temperature, Dew_Point, Humidity, Wind, Wind_Speed, Wind_Gust, Pressure, Precipitation, Condition]

        df_final = reduce(lambda left, right: pd.merge(left, right, left_index=True, right_index=True), dfs)
        output = output.append(df_final)

        filename = 'dataset/' + str(d) + '.csv'
        df_final.to_csv(filename)
        print(filename + ' completed')

    print('Scraper done!')

    output = output[['Time', 'Temperature', 'Dew_Point', 'Humidity', 'Wind', 'Wind_Speed', 'Wind_Gust', 'Pressure', 'Precipitation', 'Condition']]

    return output


def date_range(start, end):
    r = (end + datetime.timedelta(days=1)-start).days
    return [start + datetime.timedelta(days=l) for l in range(r)]

start = datetime.date(2020,6,1)
end = datetime.date(2022,6,1)
datelist = date_range(start, end)
dates = []
for date in datelist:
    new = [0,0,0]
    new[2] = str(date.day)
    new[1] = str(date.month)
    new[0] = str(date.year)
    dates.append('-'.join(new))

page = 'https://www.wunderground.com/history/daily/in/thiruvananthapuram/VOTV/date/'

df = scraper(page, dates)
```
